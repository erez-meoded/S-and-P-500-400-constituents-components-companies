```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# S&P 500 from wikipedia to csv file

url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
result = requests.get(url)
c = result.content
bs = BeautifulSoup(c,'lxml')
table = str(bs.find('table',{'class':'wikitable sortable'}))
snp500 = pd.read_html(table)[0]
snp500.columns = snp500.iloc[0]
snp500.drop(0,inplace=True)
snp500.set_index('Ticker symbol',inplace=True)
snp500.sort_index(0,inplace=True)
snp500.to_csv('snp500.csv')
snp500
```

```python
# Dow Jones from wikipedia to csv file

url = 'https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average'
result = requests.get(url)
c = result.content
bs = BeautifulSoup(c,'lxml')
table = str(bs.find('table',{'class':'wikitable sortable'}))
dj = pd.read_html(table)[0]
dj.columns = dj.iloc[0]
dj.drop(0,inplace=True)
dj.set_index('Symbol',inplace=True)
dj.sort_index(0,inplace=True)
dj.to_csv('dj.csv')
dj
```

```python
# NASDAQ100 from wikipedia to csv file

url = 'http://www.nasdaq.com/quotes/nasdaq-100-stocks.aspx'
result = requests.get(url)
c = result.content
bs = BeautifulSoup(c,'lxml')
table = str(bs.find('table',{'id':'indu_table'}))
NASDAQ100 = pd.read_html(table)[0]
NASDAQ100.columns = NASDAQ100.iloc[0]
NASDAQ100.drop(0,inplace=True)
NASDAQ100.set_index('Symbol',inplace=True)
NASDAQ100.sort_index(0,inplace=True)
NASDAQ100.to_csv('NASDAQ100.csv')
NASDAQ100
```

```python
# NASDAQ100 from nasdaq to csv file
NASDAQ100 = pd.read_csv('http://www.nasdaq.com/quotes/nasdaq-100-stocks.aspx?render=download')
NASDAQ100.set_index('Symbol',inplace=True)
NASDAQ100.sort_index(0,inplace=True)
NASDAQ100.to_csv('NASDAQ100.csv')
NASDAQ100
```
